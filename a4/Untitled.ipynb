```python
import random
hugeNumber = float("inf")

random.seed(6975)

# T
stages             = 10
startInventory     = 0
inventoryCapacity  = 40
productionCapacity = 40
num_replications  = 1

setupCost    = 3.0
variableCost = 1.0
holdingRate  = 0.5
shortageRate = 2.1
# wtf is going on h
demand = [-1000,2,2,2,2]    # The zero-th entry is unused here

    

item_production = [[0 for i in range(inventoryCapacity+1)] for j in range(stages+1)]
f = [[0 for i in range(inventoryCapacity+1)] for j in range(stages+2)]

replication_shortage_costs = []
for replication in range(num_replications):
    demand = [0] * (stages + 1)
    ordering_rates = [0] * (stages + 1)
    holding_rates = [0] * (stages + 1)
    shortage_rates = [0] * (stages + 1)
    
    for d in range(1,len(demand)):
        demand[d] = random.randint(0, 10)
    for o in range(1,len(ordering_rates)):
        ordering_rates[o] = round(random.uniform(2 , 20), 1)
    for h in range(1,len(holding_rates)):
        holding_rates[h] = round(random.uniform(5 , 15), 1)
    print(demand)
    print(ordering_rates)
    print(holding_rates)
    
    
    # change these bounds depending on the shortage rate that you're testing
    # high - [15,35], medium - [10,25], low - [5,15]
    for s in range(1,len(shortage_rates)):
        shortage_rates[s] = round(random.uniform(15 , 35), 1)
    print(shortage_rates)
    for t in range(stages,0,-1) :   
        for i in range(inventoryCapacity + 1) :
            
            # minProduction = max(0, demand[t] - i)
            minProduction = 0
            maxProduction = min(productionCapacity, inventoryCapacity - i + demand[t])

            value = hugeNumber

            for p in range(minProduction, maxProduction+1):
                productionCost = holdingCost = shortageCost = 0    

                j = i + p - demand[t]

                # if there's a shortage, it means we carry forward 0 units.
                if j < 0 :
                    j = 0

                if p < demand[t] - i:
                    # incur shortage cost
                    shortageCost += shortageRate*(demand[t] - i - p)
                    holdingCost = 0

                elif p > demand[t] - i:
                    # j is the amount left over
                    holdingCost += holdingRate*(i + p - demand[t])
                    shortageCost = 0

                productionCost = variableCost*p
                #if p > 0:
#                    productionCost += setupCost
                moveValue = holdingCost + shortageCost + productionCost + f[t+1][j]


                if moveValue < value :
                    value = moveValue
                    bestMove = p

            # End of p loop
            f[t][i] = value
            item_production[t][i] = bestMove
            #print(f)
            # print(x)

        # End of i loop
        print(f"value: {value}")
        print(f"bestMove {bestMove}\n\n")    
    # End of t loop

print("Optimal cost is " + str(f[1][startInventory]))

solutionString = "Production amounts:"
i = startInventory
for t in range(1,stages+1) :
    if i < 0:
        i = 0
    #print(t)
    solutionString += " " + str(item_production[t][i])
    #print(solutionString)
    i = i + item_production[t][i] - demand[t]

print(solutionString)
print(demand)
```
